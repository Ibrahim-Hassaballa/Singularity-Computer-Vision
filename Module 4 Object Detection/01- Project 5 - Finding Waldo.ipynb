```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('../Data/Images/WaldoBeach.jpg')
cv2.imshow('Where is Waldo?', image)
cv2.waitKey(0)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Where is Waldo?')
plt.show()

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

template = cv2.imread('../Data/Images/waldo.jpg',0)
h, w = template.shape

result = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF)
_, _, _, max_loc = cv2.minMaxLoc(result)

top_left = max_loc
bottom_right = (top_left[0] + h, top_left[1] + w)
cv2.rectangle(image, top_left, bottom_right, (0,255,0), 5)

cv2.imshow('Ahaaa, Here He is ^_^', image)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Ahaaa, Here He is ^_^')
plt.show()


cv2.destroyAllWindows()
```
