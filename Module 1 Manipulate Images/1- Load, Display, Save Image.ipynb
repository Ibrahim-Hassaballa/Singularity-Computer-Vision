# Import OpenCV Library

```python
import cv2
import matplotlib.pyplot as plt
```

# Read Image

```python
image = cv2.imread('../Data/Images/input.jpg')
cv2.imshow('Image', image)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Hello World')
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# Get Imge Shape (Width & Height)

```python
image.shape
```

```python
print("Width is: " + str(image.shape[1]) + " Pixel")
print("Height is: " + str(image.shape[0]) + " Pixel")
```

# Write Image

```python
cv2.imwrite('../Data/Output/output.jpg', image)
```
