# Crop Image

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../Data/Images/Sunflowers.jpg")

cv2.imshow("Original", img)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()

img.shape
```

```python
roi = img[200:300, 300:500]

cv2.imshow("ROI", roi)

plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.title('ROI')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
