# Convolution & Kernels

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('../Data/Images/elephant.jpg')

def apply_and_visualize_filter(img, kernel_matrix, title):
    image_after_filter = cv2.filter2D(img, -1, kernel_matrix)
    cv2.imshow(title, image_after_filter)
    cv2.waitKey(0)
    plt.imshow(cv2.cvtColor(image_after_filter, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()


cv2.imshow('Original Image', image)
cv2.waitKey(0)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.show()



# # Identity kernel
# identity_kernel = np.array([[0,0,0], [0,1,0], [0,0,0]])
# apply_and_visualize_filter(image, identity_kernel, 'Identity Filter')


# Box Blur kernel
box_kernel = np.array([[1,1,1], [1,1,1], [1,1,1]]) / 9
apply_and_visualize_filter(image, box_kernel, 'Box Blur Filter')


# Sharpen kernel
sharpen_kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
apply_and_visualize_filter(image, sharpen_kernel, 'Sharpen Filter')


# Emboss kernel
emboss_kernel = np.array([[-2,-1,0], [-1,1,1], [0,1,2]])
apply_and_visualize_filter(image, emboss_kernel, 'Emboss Filter')


# Horizontal Edge Detection kernel
horizontal_edge_detection_kernel = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
apply_and_visualize_filter(image, horizontal_edge_detection_kernel, 'Horizontal Edge Detection Filter')


# Vertical Edge Detection kernel
vertical_edge_detection_kernel = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])
apply_and_visualize_filter(image, vertical_edge_detection_kernel, 'Vertical Edge Detection Filter')


cv2.destroyAllWindows()
```

# Smoothing Images (Blur)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('../Data/Images/elephant.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.show()

# Creating our 3 x 3 kernel
kernel_3x3 = np.ones((3, 3), np.float32) / 9

# We use the cv2.fitler2D to conovlve the kernal with an image 
blurred = cv2.filter2D(image, -1, kernel_3x3)
cv2.imshow('3x3 Kernel Blurring', blurred)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('3x3 Kernel Blurring')
plt.show()

# Creating our 7 x 7 kernel
kernel_7x7 = np.ones((7, 7), np.float32) / 49

blurred2 = cv2.filter2D(image, -1, kernel_7x7)
cv2.imshow('7x7 Kernel Blurring', blurred2)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(blurred2, cv2.COLOR_BGR2RGB))
plt.title('7x7 Kernel Blurring')
plt.show()

cv2.destroyAllWindows()
```

# Advanced Bluring Techniques

```python
import cv2
import numpy as np

image = cv2.imread('../Data/Images/elephant.jpg')

# Averaging done by convolving the image with a normalized box filter. 
# This takes the pixels under the box and replaces the central element
# Box size needs to odd and positive 
blur = cv2.blur(image, (3,3))
cv2.imshow('Averaging', blur)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title('Averaging')
plt.show()

# Instead of box filter, gaussian kernel
Gaussian = cv2.GaussianBlur(image, (7,7), 0)
cv2.imshow('Gaussian Blurring', Gaussian)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blurring')
plt.show()

# Takes median of all the pixels under kernel area and central 
# element is replaced with this median value
median = cv2.medianBlur(image, 5)
cv2.imshow('Median Blurring', median)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.title('Median Blurring')
plt.show()

# Bilateral is very effective in noise removal while keeping edges sharp
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB))
plt.title('Bilateral Blurring')
plt.show()


cv2.destroyAllWindows()
```
