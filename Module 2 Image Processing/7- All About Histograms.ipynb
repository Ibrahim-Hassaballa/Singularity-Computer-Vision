# Find, Plot, Analyze

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('../Data/Images/input.jpg')
cv2.imshow('Image', image)
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Image')
plt.show()

cv2.destroyAllWindows()

color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color = col)
    plt.xlim([0,100])
    
plt.show()
```

# Histogram Equalization

```python
image = cv2.imread('../Data/Images/wiki.jpg', 0)
cv2.imshow('Image', image)
cv2.waitKey(0)

plt.imshow(image, cmap='gray')
plt.title('Image')
plt.show()

equ = cv2.equalizeHist(image)
cv2.imshow('Equalized', equ)
cv2.waitKey(0)

plt.imshow(equ, cmap='gray')
plt.title('Equalized')
plt.show()

cv2.destroyAllWindows()
```

# CLAHE (Contrast Limited Adaptive Histogram Equalization)

```python
# BAAAAAAAAAD -_-
image = cv2.imread('../Data/Images/clahe.jpg', 0)
cv2.imshow('Image', image)
cv2.waitKey(0)

plt.imshow(image, cmap='gray')
plt.title('Image')
plt.show()

equ = cv2.equalizeHist(image)
cv2.imshow('Equalized', equ)
cv2.waitKey(0)

plt.imshow(equ, cmap='gray')
plt.title('Equalized')
plt.show()

cv2.destroyAllWindows()
```

```python
# GOOOOD ^_^
image = cv2.imread('../Data/Images/clahe.jpg', 0)
cv2.imshow('Image', image)
cv2.waitKey(0)

plt.imshow(image, cmap='gray')
plt.title('Image')
plt.show()

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl_image = clahe.apply(image)

cv2.imshow('CLAHE', cl_image)
cv2.waitKey(0)

plt.imshow(cl_image, cmap='gray')
plt.title('CLAHE')
plt.show()

cv2.destroyAllWindows()
```
