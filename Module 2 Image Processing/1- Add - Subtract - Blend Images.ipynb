# Add

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../Data/Images/input.jpg', 1)

M = np.ones(image.shape, dtype = "uint8") * 200
print(M)

added = cv2.add(image, M)
cv2.imshow("Added", added)

plt.imshow(cv2.cvtColor(added, cv2.COLOR_BGR2RGB))
plt.title('Added')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Subtract

```python
import cv2
import numpy as np

image = cv2.imread('../Data/Images/input.jpg', 1)

M = np.ones(image.shape, dtype = "uint8") * 100

subtracted = cv2.subtract(image, M)
cv2.imshow("Subtracted", subtracted)

plt.imshow(cv2.cvtColor(subtracted, cv2.COLOR_BGR2RGB))
plt.title('Subtracted')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Blend

```python
import cv2
import numpy as np

image = cv2.imread('../Data/Images/ocv_logo.jpg')
image2 = cv2.imread('../Data/Images/robo.jpg')

blended = cv2.addWeighted(image, 0.7, image2, 0.3, 0)

cv2.imshow("Blended", blended)

plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.title('Blended')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
